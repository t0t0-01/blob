# Personality

### Myer-Briggs Personality Types

The Myers-Briggs Type Indicator (MBTI) is a widely recognized and influential tool for understanding personality differences. Developed by the mother-daughter duo, Katherine Briggs and Isabel Briggs Myers, the MBTI provides insights into various aspects of human behavior, preferences, and decision-making processes. By categorizing individuals into distinct personality types, the MBTI offers a framework to better comprehend ourselves and others, enabling improved communication, personal growth, and team dynamics.

The MBTI is based on Carl Jung's theories of psychological types, which propose that individuals have innate preferences in how they perceive the world, gather information, make decisions, and interact with others. According to the MBTI, these preferences fall into four dichotomies: extraversion (E) or introversion (I), sensing (S) or intuition (N), thinking (T) or feeling (F), and judging (J) or perceiving (P). By combining these preferences, the MBTI identifies 16 distinct personality types, each with its unique characteristics and strengths.

#### Extroversion/Introversion
This personality scale is about where a person draws their energy from.

Extroverts get recharged and energized from being around other people. They’re typically outgoing and action-oriented, and they can feel mentally and emotionally drained if they spend too much time on their own.

Introverts draw their energy from alone time and tend to prefer meaningful social interactions with fewer people instead of large groups. They can get fatigued and stressed if they don’t get enough time by themselves.

#### Sensing/Intuiting
This category involves the way you gather relevant information in your life.

If you lean toward sensing, you pay more attention to the tangible information you receive from your five senses. You are more likely focused on facts and concrete data that you can measure.

If, on the other hand, you rely on intuition, you’re more focused on:

abstract concepts
gut feelings
contemplating all possibilities
exploring underlying principles


#### Thinking/Feeling
This category refers to how you make decisions based on the information you gathered.

If you’re closer to the thinking side of the scale, you may prefer to make rational, logic-driven decisions that you feel are objective.

If, instead, you lean toward the feeling side of the scale, you may make decisions based on emotional and human aspects at the moment of deciding.

#### Judging/Perceiving
This scale is about how you operate on a daily basis.

If you lean toward the judging side, you may prefer structure, order, and planning in everything you do.

If you’re more of a perceiving person, you may be more flexible, adaptable, and spontaneous in your actions.

#### Limitations
Some experts have tried to identify compatible personality types based on the Myers-Briggs test. However, no exhaustive research has been conducted to validate the scales or establish a degree of compatibility. This is one of the reasons why the MBTI is typically seen as a way to entertain yourself rather than get accurate answers about your personality type.

The Myers-Briggs type indicator isn’t a validated tool to evaluate your personality type or romantic compatibility with someone else. However, it may be a fun way to question yourself and your partner in a way that fuels deep conversation about your interests and values.

The MBTI uses four scales to categorize a person’s personality based on their decision making behavior, preferences, and worldviews. It includes 16 personality types.

Some people use the test to evaluate romantic compatibility. In general, those who tend to lean on feelings seem to be more compatible with each other, even if they score differently on other scales.

But romantic compatibility is as complex as individual personality, so it’s difficult to calculate it as a mathematical formula. Still, taking the Myers-Briggs test can be a fun way to learn more about you and your partner’s interests and priorities.

This would also help with not taking it much into account, since the majority of points would be around 0. Only the really important or really bad would be taken into account.

Since we'll be doing this on several sets, let's extract the above logic into a function

```python
import pandas as pd
import numpy as np

def load_and_filter_meta(path):
    """
    Load and filter metadata from a JSON file at a given path.

    Parameters
    ----------
    path : str
        Path to the JSON file to load.

    Returns
    -------
    pandas.DataFrame
        A filtered DataFrame containing metadata from the JSON file, including one-hot encoded columns for each unique
        category value in the original metadata.

    Notes
    -----
    The function loads the JSON file into a pandas DataFrame, filters it to include only columns relevant to the analysis,
    and renames those columns. The function then applies a lambda function to replace empty category values with NaN,
    drops any rows with missing category values, and creates one-hot encoded columns for each unique category value using
    the `pandas.Series.str.get_dummies` method. The function also renames any columns that contain "&amp;" and have no
    copy with "&" only, and merges any duplicate columns resulting from the renaming. Finally, the function drops the old
    category column and returns the filtered DataFrame.

    Examples
    --------
    >>> meta_df = load_and_filter_meta("metadata.json")
    >>> print(meta_df.head())
       main_category                category   tech1  product_id  Accessories  Action  Action & Adventure  ...
    0           Baby        Activities & Toys  Gear & Equipment  0588880512            0       0                   0
    1           Baby        Activities & Toys          Swings  0594451647            0       0                   0
    2           Baby                 Apparel         Apparel  0594481813            0       0                   0
    3           Baby  Baby & Toddler Toys            Bath Toys  0594488885             0       0                   0
    4           Baby  Baby & Toddler Toys            Bath Toys  0594489032             0       0                   0
    """
    
    meta_df = pd.read_json(path, lines=True)
    
    meta_cols = ['category', 'asin', 'main_cat', 'tech1']
    meta_order = ['main_cat', "category", "tech1", "asin"]
    meta_new_cols = ["main_category", "category", "tech1", "product_id"]
    
    meta_df = meta_df.loc[:, meta_cols]
    meta_df = meta_df.reindex(meta_order, axis=1)
    meta_df = meta_df.rename(columns=dict(zip(meta_order, meta_new_cols)))
    
    # Replaces empty category values with NaN using a lambda function
    meta_df['category'] = meta_df['category'].apply(lambda x: np.nan if len(x) == 0 else x)
    
    # Drop nan values in category
    meta_df = meta_df.dropna(subset="category")
    
    # Check if df is empty to avoid error in .join
    if len(meta_df) > 0:
        # Convert list to string, then use str.get_dummies to create one-hot columns
        categories = meta_df['category'].apply("*".join).str.get_dummies(sep="*")
        
        # Rename columns that contain '&amp;' and have no copy with '&' only
        amps_cols = categories.columns[categories.columns.str.contains("&amp")]
        amps_cols_fixed = amps_cols.str.replace("&amp;", "&")
        unique = amps_cols[~amps_cols_fixed.isin(categories.columns)]
        categories.rename(lambda x: x.replace("&amp;", "&") if x in unique else x, axis=1, inplace=True)

        # Merge duplicate columns (we need to reupdate list of columsn first due to renaming above)
        amps_cols = categories.columns[categories.columns.str.contains("&amp")]
        amps_cols_fixed = amps_cols.str.replace("&amp;", "&")
        dups = list(zip(amps_cols, amps_cols_fixed))
        for wrong, fixed in dups:
            # Use clip to clip values at 1
            s = categories[[wrong, fixed]].sum(axis=1).clip(0, 1).rename(fixed)
            categories.drop([wrong, fixed], axis=1, inplace=True)
            categories = pd.concat([categories, s], axis=1)
        
        # Concatenate final categories with original df
        meta_df = pd.concat([meta_df, categories], axis=1)
    
    # Drop the old category column
    meta_df.drop(["category"], axis=1, inplace=True)

    
    return meta_df
```

Then, we need to compute the similarity of two trees. To do so, we flatten the tree, since the weights now encode the depth. 
**The question is: when comparing two different vectors, those in the same level would have the same weight. Is similarity calculation influenced by the value itself, or just how different it is? In other words, we subtract the minimum and divide by the difference of the min and max.

### Code

Now, we can start with writing the code. First of all, we define a function that would get the depths of the elements of a dictionary

```python
def get_depths(d, depth=1):
    """
    Returns a dictionary of key-depth pairs for all keys (in dot notation) in the input dictionary `d`.

    Parameters
    ----------
    d : dict
        Input dictionary whose key-depth pairs are to be computed
    depth : int
        Initial depth to start with

    Returns
    -------
    dict
        A dictionary where the keys are the original keys (in dot notation) from the input dictionary
        and the values are the corresponding depths of those keys in the input dictionary.
    """

    # Create an empty dictionary to store the results
    result = {}

    # Iterate over the key-value pairs of the input dictionary
    for key, value in d.items():

        # Add the current depth to the result dictionary for the current key
        result[key] = depth

        # If the current value is a dictionary, recursively call get_depths function to find sub-results
        if isinstance(value, dict):

            # Call get_depths function with the sub-dictionary and an incremented depth
            sub_results = get_depths(value, depth=depth + 1)

            # Iterate over the sub-results and add them to the result dictionary
            for sub_key, sub_value in sub_results.items():
                result[key + '.' + sub_key] = sub_value

    return result
```

Next, we write a function to map the dictionary to a vector. That is, we want to traverse the dictionary and assign to each key an index. This index will then be used to index the vector.

To do so, we use DFS (depth first search)

```python
def get_index_dfs(d, index=0):
    """
    Traverse a dictionary structured as a tree using Depth-First Search (DFS),
    and map the indices to the different elements of the tree.

    
    Parameters
    ----------
    d : dict
        A dictionary to be traversed.
    depth : int, optional
        The current depth of the traversal.

    Returns
    -------
    result : dict
        A dictionary mapping each element's index to a corresponding index.

    """


    # Initialize an empty dictionary to store the result
    result = {}

    # Iterate over each key-value pair in the dictionary
    for key, value in d.items():

        # Map the key's index to its depth in the tree
        result[key] = index

        # Increment the depth
        index += 1

        # If the value is another dictionary, recursively call this function
        # with the sub-dictionary and update the result dictionary
        if isinstance(value, dict):
            sub_results = get_index_dfs(value, index)

            # Iterate over each sub-result and update the result dictionary
            for sub_key, sub_value in sub_results.items():
                # Combine the parent key and sub-key with a period to create the full index
                # Map the full index to its index in the tree
                result[key + '.' + sub_key] = sub_value
                # Increment the index
                index += 1

    # Return the result dictionary
    return result
```

Next, we need to use the aforementioned functions to calculate the vector weights from the tree. Note that we also normalize the values, as discussed eariler

```python
def get_weights(tree):
    """
    Returns two dictionaries containing the weights of each concept in the tree: `normalized_concept_weights` and `concept_weights`.

    Parameters:
    -----------
    tree : dict
        A dictionary representing a tree structure.

    Returns:
    --------
    normalized_concept_weights : dict
        A dictionary containing the normalized weights of each concept in the tree.
    concept_weights : dict
        A dictionary containing the weights of each concept in the tree.
    """

    # Get the depths of each concept in the tree
    depths_of_A = get_depths(tree)
    # Determine the maximum depth in the tree
    max_depth = max(depths_of_A.values())
    # Calculate the weights of each concept in the tree based on their depth
    concept_weights = {key: value/max_depth for key, value in depths_of_A.items()}
    
    # Normalize the weights of each concept in the tree
    mx, mn = max(concept_weights.values()), min(concept_weights.values())
    normalized_concept_weights = {key: (value - mn) / (mx - mn) for key, value in concept_weights.items()}

    return normalized_concept_weights, concept_weights
```

Here, you can see the weights. They are normalized, and the deeper they are, the more weight they have

Next, we implement the similarity calculation:

```python
import numpy as np

def cos_similarity(v1, v2):
    cosine = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
    return cosine


def get_similarity(A, B, taxonomy):
    """
    Compute the cosine similarity between two vectors representing concepts in a taxonomy.

    Parameters
    ----------
    v1 : dict
        A dictionary representing the first vector of concept weights.
    v2 : dict
        A dictionary representing the second vector of concept weights.

    Returns
    -------
    float
        The cosine similarity between the two vectors.
    
    """
    # Get index for the taxonomy using DFS
    taxonomy_index = get_index_dfs(taxonomy)

    # Get the normalized and un-normalized weights for vector A
    norm_w_A, w_A = get_weights(A)
    
    # Convert the weights dictionary to a vector with a 1:1 mapping to the taxonomy index
    vector_A = np.zeros(len(taxonomy_index))
    for k, v in w_A.items():
        vector_A[taxonomy_index[k]] = v
    
    # Get the normalized and un-normalized weights for vector B
    norm_w_B, w_B = get_weights(B)
    
    # Convert the weights dictionary to a vector with a 1:1 mapping to the taxonomy index
    vector_B = np.zeros(len(taxonomy_index))
    for k,v in w_B.items():
        vector_B[taxonomy_index[k]] = v
            
            
    # Compute and return the cosine similarity between vector A and vector B
    return cos_similarity(vector_A, vector_B)
```

Now for testing purposes, we define a function that generates a random subset of a taxonomy. These will be used to calculate measures and see if they make sense:

```python
import random

def generate_random_subset(taxonomy):
    """
    Generate a random subset of a taxonomy by selecting a random number of top-level categories to include,
    and creating a dictionary containing the selected categories and their subcategories (if any).

    Parameters
    ----------
    taxonomy : dict
        A dictionary representing a taxonomy.

    Returns
    -------
    result : dict
        A dictionary containing a random subset of the original taxonomy.

    """

    # Select a random number of top-level categories to include
    num_categories = random.randint(1, len(taxonomy))

    # Select a random subset of top-level categories
    categories = random.sample(list(taxonomy.keys()), num_categories)

    # Create a dictionary containing the selected categories and their subcategories
    result = {}
    for category in categories:
        result[category] = {}
        if taxonomy[category]:
            subcategories = generate_random_subset(taxonomy[category])
            result[category] = subcategories

    return result
```
